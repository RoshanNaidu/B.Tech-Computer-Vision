```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
for i in dir(cv2):
  print(i)
```

```python
image = cv2.imread('/content/698574.jpg',cv2.IMREAD_COLOR)
```

```python
image1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image1)
```

```python
# Apply Median filter
median_filtered_img = cv2.medianBlur(image1, 5)
```

```python
plt.imshow(median_filtered_img)
plt.axis('off')
plt.title('Median Filtered Image')
```

```python
# Apply Min filter
min_filtered_img = cv2.erode(image1, np.ones((5, 5), np.uint8), iterations=1)
```

```python
plt.imshow(min_filtered_img)
plt.axis('off')
plt.title('Min Filtered Image')
```

```python
# Apply Max filter
max_filtered_img = cv2.dilate(image1, np.ones((5, 5), np.uint8), iterations=1)
```

```python
plt.imshow(max_filtered_img)
plt.axis('off')
plt.title('Max Filtered Image')
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1), plt.imshow(image1), plt.title('Original Image'), plt.axis('off')
plt.subplot(2,2,2), plt.imshow(median_filtered_img), plt.title('Median Filtered Image'), plt.axis('off')
plt.subplot(2,2,3), plt.imshow(min_filtered_img), plt.title('Min Filtered Image'), plt.axis('off')
plt.subplot(2,2,4), plt.imshow(max_filtered_img), plt.title('Max Filtered Image'), plt.axis('off')

plt.show()
```
