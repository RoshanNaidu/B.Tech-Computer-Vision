```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread("Downloads/Narayan.jpeg", cv2.IMREAD_COLOR)
```

```python
plt.title("Original BGR")
plt.axis('off')
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.title('Grayscale')
plt.axis('off')
plt.imshow(gray, cmap='gray')
```

```python
equ = cv2.equalizeHist(gray)
```

```python
plt.title("Equalized")
plt.axis('off')
plt.imshow(equ, cmap='gray')
```

```python
plt.figure(figsize=(20,20))

plt.subplot(1,3,1), plt.imshow(image), plt.title('Original'), plt.axis('off')
plt.subplot(1,3,2), plt.imshow(gray, cmap='gray'), plt.title('Grayscale'), plt.axis('off')
plt.subplot(1,3,3), plt.imshow(equ, cmap='gray'), plt.title('Equalized'), plt.axis('off')

plt.show()
```

```python
# Plotting the histograms

plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.hist(gray.flatten(), 256, [0, 256], color='red')
plt.title('Original Histogram')

plt.subplot(1, 2, 2)
plt.hist(equ.flatten(), 256, [0, 256], color='blue')
plt.title('Equalized Histogram')

plt.show()
```
